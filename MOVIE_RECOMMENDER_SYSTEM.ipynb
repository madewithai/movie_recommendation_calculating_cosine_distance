```python
def extract_genres(obj):
  genres_list = []
  for i in ast.literal_eval(obj):
    genres_list.append(i['name'])
  return genres_list
```

```python
movies['genres'] = movies['genres'].apply(extract_genres)
```

```python
def extract_keywords(obj):
  keywords_list = []
  for i in ast.literal_eval(obj):
    keywords_list.append(i['name'])
  return keywords_list
```

```python
movies['keywords'] = movies['keywords'].apply(extract_keywords)
```

```python
def extract_cast(obj):
  cast_list = []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter != 5:
      cast_list.append(i['name'])
      counter += 1
    else:
      break
  return cast_list
```

```python
movies['cast'] = movies['cast'].apply(extract_cast)
```

```python
def extract_crew(obj):
  crew_director = []
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      crew_director.append(i['name'])
      break
  return crew_director
```

```python
movies['crew'] = movies['crew'].apply(extract_crew)
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies.head(10)
```

```python
movies['tags'] = movies['overview'] + movies['keywords'] + movies['genres'] + movies['cast'] + movies['crew']
```

```python
movies_data = movies[['movie_id', 'title', 'tags']]
```

```python
movies_data.head()
```

```python
movies_data['tags'] = movies_data['tags'].apply(lambda x:" ".join(x))
```

```python
movies_data.head()
```

```python
movies_data['tags'][0]
```

```python
movies_data['tags'] = movies_data['tags'].apply(lambda x:x.lower())
```

```python
movies_data.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def text_stem(all_text):
  l = []
  for i in all_text.split():
    l.append(ps.stem(i))
  return " ".join(l)
```

```python
movies_data['tags'] = movies_data['tags'].apply(text_stem)
```

```python
movies_data['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
countvec = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = countvec.fit_transform(movies_data['tags']).toarray()
```

```python
vectors.shape
```

```python
vectors
```

```python
countvec.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_metrics = cosine_similarity(vectors)
```

```python
similarity_metrics
```

```python
similarity_metrics[0]
```

```python
def movie_recommendation(movie):
  # finding index of the movie
  index = movies_data[movies_data['title'] == movie].index[0]
  similarity_distances = similarity_metrics[index]
  recommended_list = sorted(list(enumerate(similarity_distances)), reverse=True, key=lambda x:x[1])[1:7]

  # printing the list of recommended movies
  for i in recommended_list:
    print(movies_data.iloc[i[0]].title)
```

```python
movie_recommendation('Spider-Man 3')
```
